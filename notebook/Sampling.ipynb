````python
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy.stats import truncnorm

#Domain of X
xdomain = [-3, 3]

#Multiplier Constant
M = 2.0

def pdf(x):
   """
   Probability distribution function for Random Variable X
   from which we want to sample points. Here we assume 
   we have truncated standard normal distribution in the domain of -3 to 3
   """
   return truncnorm.pdf(x, xdomain[0], xdomain[1])

def random_point_within_enveloping_region():
    """
    Return random point within the enveloping region. 
    For x we will randomly sample point between -3 and 3
    Since we are assuming uniform distribution, the height of the enveloping region
    at any x is 1/6. So for Y we randomly sample point between 0 and 1/6
    """
    #Randomly sample x from -3 to 3
    x = random.uniform(xdomain[0], xdomain[1])
    
    # probability of obtain any x is equal to 1/6. i.e. height of enveloping region
    # for any X is 1/6. 
    y = random.uniform(0, height_of_enveloping_region(x) )   
    
    return (x,y)

def height_of_enveloping_region(x):
    """Return height of enveloping region at x."""
    return M * 1.0/6.0
    
#Number of sample points to sample 
n = 100

#Creating two arrays to capture accepted and rejected points
accepted = []
rejected = []
M = 2.0

#Run this loop until we got required number of valid points
while len(accepted) < n:
    
    #Get random point
    x, y = random_point_within_enveloping_region()
    
    #If for any x if envelping region is below the distribution from which we want to sample points
    #increment the multipler constant and resample all the points. 
    if height_of_enveloping_region(x) < pdf(x):
        print "Increasing M from {0} to {1}".format(M, M+1)
        accepted = []
        rejected = []
        M += 1.0 
        continue
        
    #If y is below blue curve then accept it
    if y < pdf(x):
       accepted.append((x, y))
    #otherwise reject it. 
    else:
       rejected.append((x, y))
       
x = np.linspace(a, b, 100)
plt.plot(x, [pdf(i) for i in x], color='blue')
plt.plot(x, [height_of_enveloping_region(x)/M for i in x], color='black', ls='dashed', lw=1)
plt.plot(x, [height_of_enveloping_region(x) for i in x], color='black', ls='dashed', lw=2)
plt.plot([x[0] for x in accepted], [x[1] for x in accepted] , 'ro', color='g')
plt.plot([x[0] for x in rejected], [x[1] for x in rejected] , 'ro', color='r')
plt.show()  

#Calculate expected value for the truncated standard normal distribution
approxMean =  sum([x[0] for x in accepted])/len(accepted)
print "Expected Mean = ", 0, pdf(0)
print "Approximated Mean = ", approxMean, pdf(approxMean)
print "Approximated Variance = ", sum([(x[0] - approxMean)**2 for x in accepted])/(len(accepted)-1)
````

````python

[caption id="attachment_2367" align="alignleft" width="263"]<a href="https://ragrawal.files.wordpress.com/2014/12/circle.png"><img class="wp-image-2367 size-full" src="https://ragrawal.files.wordpress.com/2014/12/circle.png" alt="circle" width="263" height="256" /></a> Figure (A): Unit Circle[/caption]


Let's assume you want to calculate area of the circle shown in the figure but without any formula. One way to solve this problem is using **rejection sampling** i.e. we sample points and reject points that are outside of the circle. The intuition behind it very simple. As shown in figure A, consider a square box perfectly enclosing the circle. Now the probability of selecting any point within this square box is uniform. However what's the probability of a randomly selected point to be within the circle ? It's equal to $latex \frac{Area(Circle)}{Area(Box)}$. If we randomly draw points from the box and calculate percentage of points that fall within the circle  we will notice that it is equal to percentage area of the circle i.e. $latex \frac{\#PointsWithinCircle}{\#Samples} = \frac{Area(Circle)}{Area(Box)}$. Below is a python code that computes area by drawing 1000 points. Based on 1000 points the estimated area of the circle to be 0.775 sq units (actual area is 0.785).  

```python
import matplotlib.pyplot as plt
import math

def dist(x, y, center_x = 0.5, center_y = 0.5):
#Check if the point is within the circle or not.
return math.sqrt((center_x - x) ** 2 + (center_y - y) ** 2)

#Coordinates that are within the circle
accepted = []

#Coordinates that outside of the circle
rejected = []

#Number of points we want to sample
n = 1000

#Radius of the circle
radius = 0.5

for i in range(n):

#Radomly select a coordinate
x = random.random()
y = random.random()

#Calculate distance from the center of the circle
distance = dist(x,y, 0.5, 0.5)

#if point is within the circle then accept it
#otherwise reject it.
if distance &lt;= radius:
accepted.append((x,y))
else:
rejected.append((x,y))

#Propotion of the points that are within the circle.
print "Percentage of Points Within Circle = ", len(accepted)*1.0/(len(accepted) + len(rejected))

#Plot the circle and points
circle = plt.Circle((0.5, 0.5), radius=0.5, fc='y')
plt.gca().add_patch(circle)
plt.plot([x[0] for x in accepted], [x[1] for x in accepted] , 'ro', color='g')
plt.plot([x[0] for x in rejected], [x[1] for x in rejected] , 'ro', color='r')
plt.axis('scaled')
plt.show()
```

<a href="https://ragrawal.files.wordpress.com/2014/12/circlewithpoints.png"><img class="aligncenter size-full wp-image-2372" src="https://ragrawal.files.wordpress.com/2014/12/circlewithpoints.png" alt="CircleWithPoints" width="263" height="256" /></a>

Now's let's extend this idea to probability distributions. In case of Bayesian estimator, we use sampling technique to estimate distribution parameters such as mean or variance. Let's see how the above extends to this problem. For simplicity, let's assume we want to estimate mean of a gaussian distribution. 
````
